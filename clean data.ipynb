```python
val = df["Actual_Spent_USD"].max()/6
```

```python
bins=[0,val,val*2,val*3,val*4,val*5,val*6]
labels=[">1000k","100-500k","500-5m","1m-6m","5m-10m","10m"]
```

```python
val*4
```

```python
df["cust_bucket"]=pd.cut(df["Actual_Spent_USD"],bins=bins, labels=labels, include_lowest=True)
```

```python
df 
```

```python
Actual_Spent_USD = df.groupby("cust_bucket", observed=True)["Actual_Spent_USD"].sum()
```

```python
Actual_Spent_USD
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.bar(labels, Actual_Spent_USD, color="skyblue")
plt.title("totall actual spent by cost bucket")
plt.xlabel("cost_bucket")
plt.ylabel("Actual_Spent_USD (USD)")
plt.xticks(rotation=45)
plt.grid(True, axis="y")
plt.tight_layout()
plt.show()
```
